# Import libraries

```python
import pandas as pd

import utils_finbert as uf
```

```python
INPUT_PATH = "../data/v2/bsky_emotion_filtered_posts.csv"
OUTPUT_PATH = "../data/v2/bsky_finbert_posts.csv"
SUMMARY_PATH = "../data/v2/bsky_finbert_summary.csv"
TEXT_COLUMN = "Clean_Comment"
```

# Load emotion-filtered posts

```python
posts_df = pd.read_csv(INPUT_PATH)
posts_df.shape
```

# FinBERT sentiment

```python
finbert_classifier = uf.load_finbert_classifier(
    model_name="ProsusAI/finbert",
    device=-1  # use 0 if you have GPU
)

finbert_df = uf.add_finbert_columns(
    posts_df,
    classifier=finbert_classifier,
    text_column=TEXT_COLUMN,
    batch_size=32,
    progress_every=1000,
)

finbert_summary = uf.summarize_finbert(finbert_df)
display(finbert_summary)
```

# Save results

```python
uf.save_finbert_results(finbert_df, OUTPUT_PATH)
uf.save_finbert_results(finbert_summary, SUMMARY_PATH)
```
